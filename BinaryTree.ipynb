```python
class TreeNode:
    def __init__(self, val) -> None:
        self.val = val
        self.left = None
        self.right = None 
```

```python
"""
     1
    / \
   2   3
  / \ 
 4   5
"""
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
```

```python
def height(root):
    if not root:
        return 0
    return 1 + max(height(root.left), height(root.right))
```

```python
print(height(root))
```

```python
def preorder(root):
    if not root:
        return
    print(root.val)
    preorder(root.left)
    preorder(root.right)
```

```python
preorder(root)
```

```python
def inorder(root):
    if not root:
        return
    inorder(root.left)
    print(root.val)
    inorder(root.right)
```

```python
inorder(root)
```
