```python
class TreeNode:
    def __init__(self, val) -> None:
        self.val = val
        self.left = None
        self.right = None 
```

```python
"""
     1
    / \
   2   3
  / \ 
 4   5
"""
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)

s_root = TreeNode(1)
s_root.left = TreeNode(2)
s_root.right = TreeNode(2)
s_root.left.left = TreeNode(3)
s_root.left.right = TreeNode(4)
s_root.right.left = TreeNode(4)
s_root.right.right = TreeNode(3)
```

```python
"""
(the number of nodes along the longest path from root to leaf).
"""
def height(root):
    if not root:
        return 0
    return 1 + max(height(root.left), height(root.right))

print(height(root))
```

```python
def preorder(root):
    if not root:
        return
    print(root.val)
    preorder(root.left)
    preorder(root.right)
    
preorder(root)
```

```python
def inorder(root):
    if not root:
        return
    inorder(root.left)
    print(root.val)
    inorder(root.right)
    
inorder(root)
```

```python
def postorder(root):
    if not root:
        return
    postorder(root.left)
    postorder(root.right)
    print(root.val)

postorder(root)
```

```python
"""
Find Maximum Depth
(the number of edges along the longest path from root to leaf).
"""

def Depth(root):
    if not root:
        return -1
    left = Depth(root.left)
    right = Depth(root.right)
    return 1 + max(left, right)

Depth(root)
```

```python
"""
Given a binary tree, return a list of lists — 
each inner list contains the values of nodes at that depth, 
in left-to-right order.
"""
from collections import deque

def level_order(root):
    if not root:
        return []
    queue = deque([root])
    result = []
    while queue:
        level_size = len(queue)
        level = []
        for _ in range(level_size):
            node = queue.popleft()
            level.append(node.val)
            if node.left:  queue.append(node.left)
            if node.right: queue.append(node.right)
        result.append(level)
        
    return result

level_order(root)
```

```python
"""
Given a binary tree, check whether it is 
symmetric around its center — i.e. it's a mirror image of itself.
"""
def is_mirror(left, right):
    if left is None and right is None:
        return True
    if not left or not right:
        return False
    if not left.val == right.val:
        return False
    
    L = is_mirror(left.left, right.right)
    R = is_mirror(left.right, right.left)
    return L and R

    
def is_symmetric(root):
    if not root: 
        return True
    return is_mirror(root.left, root.right)

is_symmetric(s_root)
```
